```
#carregando a tabela gerada com a quantidade de ocorrÃªncias por estabelecimento no Brasil de 2016-2020
path = '/FileStore/tables/test_parquet/apac_brasil/estabelecimentos/comparacao__1_.csv'


infer_schema = "true"
first_row_is_header = "true"
delimiter = ","
file_type = "csv"

coduni_df = spark.read.format(file_type) \
      .option("inferSchema", infer_schema) \
      .option("header", first_row_is_header) \
      .option("sep", delimiter) \
      .load(path)

coduni_df.display()
```

```
print(coduni_df.count())
```

```
arrayCod = coduni_df.select("PA_CODUNI").rdd.map(lambda r: r[0]).collect() 
arrayCod
```

```
#importando os dados dos estabelecimentos registrados no DataSus
import os
import pyspark.sql.functions as sql_fun

path = '/FileStore/tables/test_parquet/apac_brasil/estabelecimentos/cnes_estabelecimentos.csv'


infer_schema = "true"
first_row_is_header = "true"
delimiter = ","
file_type = "csv"

estab_df = spark.read.format(file_type) \
      .option("inferSchema", infer_schema) \
      .option("header", first_row_is_header) \
      .option("sep", delimiter) \
      .load(path)

estab_df.display()
```

```
estab_df = estab_df.orderBy('CO_CNES', ascending=True)
estab_df.display()
```

```
#relacionando os estabelecimentos e cruzando as informaÃ§Ãµes com as anÃ¡lises geradas durante anÃ¡lise dos dados
filtered_df = estab_df.filter(estab_df.CO_CNES==arrayCod[0])
filtered_df.display()
```
