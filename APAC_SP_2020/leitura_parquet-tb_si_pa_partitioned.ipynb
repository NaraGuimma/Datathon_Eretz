```
#preparação dos dados originais, filtrando-os por CID10 igual a L20 e suas subcategorias
import os
import pyspark.sql.functions as sql_fun

filenames_and_paths = []
directory = "/dbfs/FileStore/tables/test_parquet/SP_2020"
dir="/FileStore/tables/test_parquet/SP_2020"

infer_schema = "true"
first_row_is_header = "true"
delimiter = ","
file_type = "parquet"

for filename in os.listdir(directory):
  if filename.endswith(".parquet"):
    file_path=os.path.join(dir, filename)
    filenames_and_paths.append(file_path)

# print(file_path)    
# print(filenames_and_paths)    
     
df = spark.read.format(file_type) \
      .option("inferSchema", infer_schema) \
      .option("header", first_row_is_header) \
      .option("sep", delimiter) \
      .load(filenames_and_paths[0])
result = df.filter(df.PA_CIDPRI.contains('L20'))
  
for i in range(1, len(filenames_and_paths)):
  df = spark.read.format(file_type) \
        .option("inferSchema", infer_schema) \
        .option("header", first_row_is_header) \
        .option("sep", delimiter) \
        .load(filenames_and_paths[i])
  df_2 = df.filter(df.PA_CIDPRI.contains('L20'))
  print(filenames_and_paths[i])
  
  result = result.unionByName(df_2)
  
result.display()
```
